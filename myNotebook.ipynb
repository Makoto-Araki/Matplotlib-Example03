```python
# ライブラリのインポート
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
import pandas as pd

# CSVファイルから読込
sales = pd.read_csv('C:\\Users\Owner\Matplotlib-Exercises3\\FILE1.csv')
```

```python
# データフレームの内容
sales
```

```python
# X軸とY軸の値および各部分円の色
X = sales['商品名']
Y = sales['売上高']
C = ['r', 'c', 'b', 'm', 'y', 'g']

# グラフ装飾
plt.title('各商品売上', fontsize=20)
plt.rcParams['font.size'] = 12.0

# グラフ描画
plt.pie(Y, labels=X, counterclock=True, autopct="%1.1f%%", colors=C)
plt.show()
```
